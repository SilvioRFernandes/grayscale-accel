## Importando Bibliotecas

```python
from PIL import Image
```

### Lendo arquivo referente a imagem em questão / pegando largura e altura da imagem recebida

```python
arquivo = Image.open("../images/lena.jpg")
imagem = arquivo.load()
largura, altura = arquivo.size
```

### Criando uma nova imagem

```python
NovaIgagem = Image.new('RGB', (largura, altura))
setarPixel = NovaIgagem.load()
```

### Funcao de Conversao para receber os pixels da imagem anterior e convertelos para a nova

```python
def ConverteEmCinza(vet):
    R = vet[0] * 0.299
    G = vet[1] * 0.587
    B = vet[2] * 0.114

    YIQ = R + G + B
    return YIQ
```

### Realizando conversao pixel a pixel da antiga imagem e setando na nova
* OBS: Os resultados obtidos pelas conversoes, de acordo com a documentação da biblioteca PIL, não podem ser de outro tipo se não inteiro.

```python
import time

start_time = time.time()
for x in range(largura):
    for y in range(altura):
       resultado = ConverteEmCinza(imagem[x, y])
       setarPixel[x, y] = (int(resultado), int(resultado), int(resultado))

stop_time = time.time()
sw_exec_time = stop_time-start_time
print('Time to software calculation: ',sw_exec_time)
```

### Gerando um novo arquivo para a nova imagem(resultante)

```python
NovaIgagem.save('../images/LenaCinza.jpg')
```
